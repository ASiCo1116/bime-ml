```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
def rescale(arr):
    _min = arr.min()
    _max = arr.max()
    return (arr - _min) / (_max - _min)
```

```python
data = loadmat('04HW2_noisy.mat')
data['X'].shape
```

```python
plt.imshow(data['X'][:, 10].reshape(28, 20), cmap='gray')
plt.show()
```

### 3(a)

```python
from sklearn.decomposition import PCA

dataT = data['X'].transpose()
pca = PCA(100).fit(dataT)
for i in [10, 121, 225, 318, 426]:
    print(i)
    new_face = 0
    img = data['X'][:, i].reshape(28, 20)
    plt.imshow(img, cmap='gray')
    plt.show()
    value = pca.transform(img.reshape(-1, 560))
    for idx, evtr in enumerate(pca.components_):
        new_face += value[0][idx] * evtr
        if idx == 10:
            break
        
    new_face = new_face.reshape(28, 20)
    plt.imshow(new_face, cmap='gray')
    plt.show()
```

### 3(b)

```python
for i in [10, 121, 225, 318, 426]:
    print(i)
    new_face = 0
    img = data['X'][:, i].reshape(28, 20)
    plt.imshow(img, cmap='gray')
    plt.show()
    value = pca.transform(img.reshape(-1, 560))
    for idx, evtr in enumerate(pca.components_):
        new_face += value[0][idx] * evtr
        if idx == 2:
            break
        
    new_face = new_face.reshape(28, 20)
    plt.imshow(new_face, cmap='gray')
    plt.show()

for i in [10, 121, 225, 318, 426]:
    print(i)
    new_face = 0
    img = data['X'][:, i].reshape(28, 20)
    plt.imshow(img, cmap='gray')
    plt.show()
    value = pca.transform(img.reshape(-1, 560))
    for idx, evtr in enumerate(pca.components_):
        new_face += value[0][idx] * evtr
        if idx == 30:
            break
        
    new_face = new_face.reshape(28, 20)
    new_face = rescale(new_face)
    plt.imshow(new_face, cmap='gray')
    plt.show()
```
